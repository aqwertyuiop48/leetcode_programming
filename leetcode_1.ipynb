```python
'''
Add the Secret to Colab
- Open your Colab notebook.
- Click the Key icon on the left-hand toolbar.
- Click "Add new secret".Set the Name (e.g., LEETCODE_USERNAME) and paste your token into the Value field.
- Important: Toggle the "Notebook access" switch to ON for that specific secret.
'''

!npm install -g leetcode-tools/leetcode-cli
!git clone https://github.com/aqwertyuiop48/leetcode_programming.git

from google.colab import userdata
import os

# 1. Pull from Colab's Secret storage
try:
    user_name = userdata.get('LEETCODE_USERNAME')
    user_cookie = userdata.get('LEETCODE_TOKEN')
except userdata.SecretNotFoundError:
    print("Error: Secrets not found. Please add them to the 'Keys' tab on the left.")
    user_name = user_cookie = None

if user_name and user_cookie:
    # 2. Assign to OS environment so the bash '!' commands can see them
    os.environ['LEETCODE_USERNAME'] = user_name
    os.environ['LEETCODE_TOKEN'] = user_cookie

    print(f"Attempting login for: {user_name}")

    # 3. Run the login using the same printf logic that worked for you
    !export PATH=$PATH:$(npm config get prefix)/bin && \
     printf "%s\n%s\n" "$LEETCODE_USERNAME" "$LEETCODE_TOKEN" | leetcode user -c && \
     leetcode user
```
